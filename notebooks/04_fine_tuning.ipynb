```python
X_train = pd.DataFrame(df_X_train)
X_test = pd.DataFrame(df_X_test)
y_train = pd.DataFrame(df_y_train).values.ravel()
y_test = pd.DataFrame(df_y_test).values.ravel()
```

# Hyperparameter Tuning

## Tuning

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
param_grid = {'alpha': np.logspace(-2, 3, 10)}
ridge_grid = GridSearchCV(Ridge(), param_grid, cv=5)
ridge_grid.fit(X_train_scaled, y_train)
```

```python
best_ridge_model = ridge_grid.best_estimator_
print(f"✅ Best Alpha: {ridge_grid.best_params_['alpha']}")
y_pred = best_ridge_model.predict(X_test_scaled)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

```python
y_pred = best_ridge_model.predict(X_test_scaled)
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted_Values'])
y_pred_df.to_csv('/content/drive/My Drive/Professional/Portfolio/test_split/y_pred.csv', index=False)
joblib.dump(best_ridge_model, '/content/drive/My Drive/Professional/Portfolio/test_split/best_ridge_model.pkl')
```
