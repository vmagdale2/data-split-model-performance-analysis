```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# Model Evaluation

```python
df_y_pred = np.ravel(df_y_pred)
df_y_test = np.ravel(df_y_test)
```

```python
residuals = df_y_test - df_y_pred

plt.figure(figsize=(8, 5))
sns.residplot(x=df_y_pred, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 2})
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_y_test, df_y_pred, alpha=0.5)
plt.plot([min(df_y_test), max(df_y_test)], [min(df_y_test), max(df_y_test)], color='red', linestyle='--')
plt.title('Predicted vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(residuals, kde=True, bins=30)
plt.axvline(residuals.mean(), color='red', linestyle='--', label='Mean')
plt.title('Error Distribution (Residuals)')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import joblib

# Load the trained model
model = joblib.load('/content/drive/My Drive/Professional/Portfolio/test_split/best_ridge_model.pkl')
print("✅ Model successfully loaded.")
```

```python
try:
    feature_importance = pd.Series(model.coef_, index=df_X_train.columns).sort_values(key=abs, ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature_importance.values, y=feature_importance.index)
    plt.title('Feature Importance (Linear Model)')
    plt.xlabel('Coefficient Value (Strength of Impact)')
    plt.show()
except AttributeError:
    importances = model.feature_importances_
    features = df_X_train.columns

    plt.figure(figsize=(10, 6))
    sns.barplot(x=importances, y=features)
    plt.title('Feature Importance (Tree-Based Model)')
    plt.xlabel('Importance Score')
    plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(df_y_test, df_y_pred))
mae = mean_absolute_error(df_y_test, df_y_pred)
r2 = r2_score(df_y_test, df_y_pred)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

# Model Evaluation Results

## 📊 Overview
This document provides an analysis of the model's performance metrics and key visualizations. The Ridge Regression model's improved performance reflects successful tuning and scaling adjustments.

---

## 📈 Performance Metrics
| **Metric**                | **Value**        |
|---------------------------|------------------|

### 🔎 **Interpretation:**
✅ **R² = 0.82** — A strong fit, indicating that 82% of the variance in the target variable is explained by the model. This is a solid result for a housing price prediction model.

✅ **RMSE = 32,644** — Indicates an average error of ~32K per prediction. While this is reasonable given housing price variability, further refinements could reduce this.

✅ **MAE = 19,036** — Represents the average absolute error, suggesting most predictions are within ~$19K of the true value.

---

## 📊 Visualizations Analysis

### **1. Residual Plot**
- **Trend Observation:** The red trendline shows a slight upward curve, suggesting minor **heteroscedasticity** (increasing variance as predictions grow).
- **Interpretation:** While this is common in housing data due to higher-value properties having more variability, it may suggest further feature engineering could improve accuracy at the upper range.

### **2. Predicted vs. Actual Values Plot**
- **Trend Observation:** The points are tightly clustered along the diagonal red line, indicating strong predictive performance.
- **Interpretation:** Outliers are minimal, reinforcing that the model generalizes well.

### **3. Error Distribution Plot**
- **Trend Observation:** Errors are roughly centered around zero, with a near-normal distribution.
- **Interpretation:** This confirms the model’s errors are well-balanced, with minimal systematic bias.

### **4. Feature Importance Analysis**
- **Observation:** The model assigns significant weight to various features, with some coefficients reaching extreme positive or negative values.
- **Interpretation:** Feature importance reveals key variables driving model performance. The dense visualization suggests some redundancy; applying feature selection methods like **Recursive Feature Elimination (RFE)** may enhance stability.
