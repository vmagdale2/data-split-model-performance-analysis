```python
linear_model = LinearRegression()
ridge_model = Ridge(alpha=10)
lasso_model = Lasso(alpha=0.5, max_iter=5000)
elasticnet_model = ElasticNetCV(l1_ratio=[0.1, 0.3, 0.5, 0.7, 1.0], alphas=[1, 2, 5, 10, 20, 50], max_iter=50000, cv=5, tol=1e-6)
```

```python
y_pred_linear = linear_model.predict(X_test_scaled)
y_pred_ridge = ridge_model.predict(X_test_scaled)
y_pred_lasso = lasso_model.predict(X_test_scaled)
y_pred_elasticnet = elasticnet_model.predict(X_test_scaled)
```

```python
def evaluate_model(model_name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{model_name} - MSE: {mse:.2f}, MAE: {mae:.2f}, R² Score: {r2:.2f}')

evaluate_model('Linear Regression', y_test, y_pred_linear)
evaluate_model('Ridge Regression', y_test, y_pred_ridge)
evaluate_model('Lasso Regression', y_test, y_pred_lasso)
evaluate_model('ElasticNetCV', y_test, y_pred_elasticnet)
```
